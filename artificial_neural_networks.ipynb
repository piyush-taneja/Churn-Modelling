```python
# Part-1 Data Preprocessing

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
dataset
```

```python
X
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X=X[: ,1:]
```

```python
X
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split  # we replaced cross_validation by model_selection due to spyder updation which has model_selection as value 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# Part 2 - Now let's make the ANN!

#Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Inintialising the ANN
classifier = Sequential()
```

```python
#Adding the input layer and first hidden layer
classifier.add(Dense(output_dim=6 , init='uniform' , activation='relu' , input_dim=11))
#output_dim is used to tell no. of hidden cells in first layer i.e is calculated by taking average of (no. of input layer + no. of output layer)/2
# We need to use input_dim only for the first hidden layer after that all the others hidden layers will automatically understand about it. It tells about no. of input variables in hidden layer
# init='uniform' is used to initialize the weights uniformly.
# Here we use activation function whose value is 'relu' i.e rectifier activation function. 
```

```python
# Adding second Hidden layer
classifier.add(Dense(output_dim=6 , init='uniform' , activation='relu')) # Not required to write input_dim as written already for first layer.
```

```python
classifier
```

```python
# Adding the output layers
classifier.add(Dense(output_dim=1 , init='uniform' , activation='sigmoid' )) # Here since we are finding the o/p layer so value of activation is 'sigmoid' as sigmoid func is used at output side.
```

```python
#Compiling the ANN
classifier.compile(optimizer='adam' , loss='binary_crossentropy' , metrics=['accuracy'])
```

```python
#Fitting the ANN to the Training set
classifier.fit(X_train , y_train , batch_size=10 , nb_epoch=100)
```

```python
# Part 3 - Making the predictions and Evaluating the model
 
#Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5) # Here this is done to convert probability into true or false i.e whether proba. of customer leaving bank is>0.5 then true else false
# Making the confusion Matrix
from sklearn.metrics import confusion_matrix  #Here confusion_matrix is a function and not a class as class has capitalletters at the beginning.
cm = confusion_matrix(y_test , y_pred)
```
